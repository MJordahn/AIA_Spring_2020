## Outline:

#### Explain the image segmentation problem:
* A (grayscale) image is a function $I:\Omega \to \mathbb{R}$ where $\Omega \subseteq \mathbb{R}^2$
* Given a labelset $L= \{1,...,n\}$ of $n$ labels, at partition of an image is a function $f: \Omega \to L$. That is each pixel is mapped to a label.
* Naive approach: Thresholding using the pixel intensities. However the location of the pixels relative to oneanother is not taken into account then, and e.g. textures will not be segmented correctly.

#### Supervised approach:
* A supervised approach will be adopted. Thus we will extract features from an image with a known segmentation. Using k-nearest neighbour we can now make inference about the segmentation of a test-image.

#### Extract features:
1. Method 1: We examine the image, and its derivatives. Consider an image $I$
    *  Let L denote $I*g(t)$, where $g(t)$ denotes a gaussian with $\sigma^2 = t$.
    *  We now construct a volume $F: \Omega \times [1, 2, ..., 15] \to \mathbb{R}$, by letting: $$F = [L, L_{x}, L_{y}, L_{xx}, L_{yy}, ..., L_{xyyy}, L_{yyyy}]$$
    i.e. finding all derivatives of $I*g(t)$ up to an order of 4.
	*  Note that the n'th derivative of an image convoluted with a gaussian equals the image convoluted with the n'th derivative of a gaussian. (PROPERTY OF CONVOLUTIONS).
	*  Note that we now can associate a 15 dimensional feauture vector to each pixel.
    *  Edgecases are handled by padding with the reflection.
2. Method 2: A featurevector for each pixel can be created using the $9x9$-patch surrounding the pixel. Edgecases are handled using reflection padding.

#### Classification:
1. The classification of each feature vector can be done using knn. With the euclidean distance measure. This method is however not found to be robust.
2. Instead k-means clutering can be performed on the features of the training image. The distribution of labels within each cluster can then be used to assign a probability for a certain label for a feature vector of the test image. This will result in a probability image. This is the method that will be used in the following.

#### Key take aways:
1. For the "A" images a high quality segmentation can be gotten using the gaussian features and low smoothing.
2. For the "B" images a medium quality segmentation can be gotten using the gaussian features and high smoothing.
3. For the "C" images it is not possible to get a good segmentation using gaussian features. However a patch-based smoothing procedure produces a high quality output. This can be due to the greatly increased amount of features.
4. Note: The use of patch-based image segementation reduces size of the image. This can be avoided with reflection padding.
5. On the bone images the gaussian feature extraction proved best.

### Define helper functions:

```python
# Import helpers:
import cv2
import numpy as np
import matplotlib.pyplot as plt
import scipy
import sklearn
from scipy import ndimage
import math
from feature_based_segmentation import *
import warnings
warnings.filterwarnings('ignore')

def importPictures(name):
    if name=='A':
        img_train = cv2.imread('Ex3_data_code/data/train_A_image.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_background = cv2.imread('Ex3_data_code/data/test_A_background.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_foreground = cv2.imread('Ex3_data_code/data/test_A_foreground.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_comb = cv2.imread('Ex3_data_code/data/test_A_image.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_truth = cv2.imread('Ex3_data_code/data/test_A_ground_truth.png', cv2.IMREAD_GRAYSCALE).astype(float)
    elif name=='B':
        img_train = cv2.imread('Ex3_data_code/data/train_B_image.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_background = cv2.imread('Ex3_data_code/data/test_B_background.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_foreground = cv2.imread('Ex3_data_code/data/test_B_foreground.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_comb = cv2.imread('Ex3_data_code/data/test_B_image.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_truth = cv2.imread('Ex3_data_code/data/test_B_ground_truth.png', cv2.IMREAD_GRAYSCALE).astype(float)
    elif name=='C':
        img_train = cv2.imread('Ex3_data_code/data/train_C_image.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_background = cv2.imread('Ex3_data_code/data/test_C_background.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_foreground = cv2.imread('Ex3_data_code/data/test_C_foreground.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_comb = cv2.imread('Ex3_data_code/data/test_C_image.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_truth = cv2.imread('Ex3_data_code/data/test_C_ground_truth.png', cv2.IMREAD_GRAYSCALE).astype(float)
    elif name=='bone':
        img_train = cv2.imread('Ex3_data_code/data/bone_train.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_comb = cv2.imread('Ex3_data_code/data/bone_test.png', cv2.IMREAD_GRAYSCALE).astype(float)
        img_test_truth = cv2.imread('Ex3_data_code/data/bone_label.png', cv2.IMREAD_GRAYSCALE).astype(float)
        dici = {'train': img_train,
                'test1': img_test_comb,
                'true': img_test_truth}
    if name!='bone':
        dici = {'train': img_train,
                'test1': img_test_background,
                'test2': img_test_foreground,
                'test3': img_test_comb,
                'true': img_test_truth}
    return dici

# Feature extraction is made for all images:
def featureDict(dic):
    feature_dict = {}
    for i in dic:
        if i != 'true':
            im = get_gauss_feat_im(dic[i], 2.5)
            dim = im.shape
            feature_dict[i] = np.reshape(im, (dim[0]*dim[1], dim[2]))
        else:
            im = dic[i]
            dim = im.shape
            feature_dict[i] = np.reshape(im, (dim[0]*dim[1]))
    return feature_dict, dim[0]

def patchFeatureDict(dic):
    feature_dict = {}
    for i in dic:
        if i != 'true':
            im = ndim2col(dic[i], [9, 9])
            d= math.sqrt(im.shape[1])
            feature_dict[i] = im.transpose()
        else:
            im = dic[i]
            dim = im.shape
            feature_dict[i] = np.reshape(im[4:-4,4:-4], ((dim[0]-8)*(dim[1]-8)))
    return feature_dict, int(d)

# Get probability matrix:
def probabilityMatrix(predictions, centers, pic_label):
    probs = np.zeros((centers, 2))
    for i, item in enumerate(predictions):
        if pic_label[i] == 0:
            probs[item][0] += 1
        else:
            probs[item][1] += 1

    for i, item in enumerate(probs):
        probs[i] = item/sum(item)
    return(probs)

def kmeansTrainer(feature_dict, subsample=10000, centers=500):
    n, _ = feature_dict['train'].shape

    # subsampling:
    idx = np.random.permutation(np.arange(n))[:subsample]
    pic_subsample = np.take(feature_dict['train'], idx, axis=0)
    pic_label = np.take(feature_dict['true'], idx, axis=0)

    # fit model:
    model = sklearn.cluster.KMeans(centers)
    model.fit(pic_subsample)

    # Make predictions from subsample:
    predictions_train = model.predict(pic_subsample)

    # Create probability matrix:
    probs = probabilityMatrix(predictions_train, centers, pic_label)

    return(model, probs)
```

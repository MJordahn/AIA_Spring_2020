```python
import numpy as np
import matplotlib.pyplot as plt
from make_data import make_data

def init_net_weights(neurons_in_layers):
    W = []
    for i in range(len(neurons_in_layers)-1):
        weights_in_layer = (neurons_in_layers[i+1], neurons_in_layers[i]+1)
        W = W + [np.random.normal(size=weights_in_layer)*np.sqrt(2)/np.sqrt(neurons_in_layers[i]+1)]
    return W

def forwardpass(x,W):
    H = []
    h = x
    # forward pass    
    for i in range(len(W)):
        h = np.r_[h, np.ones((1,h.shape[1]))]
        H = H + [h]
        z = W[i].dot(h) # linear part
        h = np.maximum(z, 0) # activation

    return H, z

def backpass(x, t, W, z, H):
    gradients = []
    z = softmax(z)
    delta = z-t # Compute last layer delta first
    for i in range(len(W)-1, -1, -1):
        gradients = [delta.dot(H[i].T)/x.shape[1]] + gradients # We compute all errors at same time so have to divide by number of inputs
        delta = (H[i]>0)*W[i].T.dot(delta) # Compute delta for layer l-1
        delta = delta[:-1] # Remove bias
        
    loss = -np.log(z[t.astype(bool)]).sum()
    return gradients, loss, z

def softmax(predictions):

    cq = np.exp(predictions)
    cq = (cq/cq.sum(axis=0))
    return cq

def update(W,G,eta):
    # updating network
    for i in range(len(W)):
        W[i] = W[i] - eta*G[i]
    return W

layers = [2,4,2]
weights = init_net_weights(layers)
X, T, test, dim = make_data(example_nr=1, n=200)
X = X.T
T = T.T
test = test.T
outputs = []
L = []
eta = 0.001
for i in range(10000):
    H, z = forwardpass(X, weights)
    gradients, loss, z = backpass(X, T, weights, z, H)
    L = L + [loss]
    weights = update(weights, gradients, eta)
#plt.plot(*range(1000), L)
classes = []
for a_class in zip(z.T,X.T):
    prediction = np.argmax(a_class[0])
    if prediction == 0:
        plt.scatter(a_class[1][0],a_class[1][1], color='red')
    elif prediction == 1:
        plt.scatter(a_class[1][0],a_class[1][1], color='blue')
plt.show()
```

```python
import cv2
import math
from matplotlib import pyplot as plt
im = cv2.imread("week_1_data/fibres_xcth.png", cv2.IMREAD_GRAYSCALE)
im = im.astype(float)
```

```python
plt.imshow(im)
plt.show()
```

```python
from scipy import ndimage
from scipy.stats import norm
import numpy as np
sigma = 3
t = sigma**2
filt = np.array([*range(-3*sigma,sigma*3+1)])
g = np.exp(-( (filt)**2 / ( 2.0 * sigma**2 ) ) )/np.sqrt(sigma**2*2*math.pi)
g = np.array([g])
im_g = ndimage.convolve(im, g)
g = g.transpose()
im_g = ndimage.convolve(im_g, g)
```

```python
plt.imshow(im_g)
plt.show()
plt.plot(g)
plt.show()
```

```python
from scipy import ndimage
from scipy.stats import norm
import numpy as np
import math
sigma = 3
t = sigma**2
array = []
for sig_1 in range(-sigma*3, sigma*3+1):
    row = []
    for sig_2 in range(-sigma*3, sigma*3+1):
        row.append((sig_2,sig_1))
    array.append(row)
g_2d = []
for row in array:
    g_2d_row = []
    for element in row:
        g_2d_row.append(np.exp(-((element[0]**2+element[1]**2) / ( 2.0 * sigma**2 )))/(2*math.pi*sigma**2))
    g_2d.append(g_2d_row)
g_2d = np.array(g_2d)
im_g_2d = ndimage.convolve(im, g_2d)
```

```python
plt.imshow(im_g_2d)
plt.show()

plt.imshow(im_g_2d-im_g)
plt.show()
plt.imshow(g_2d)
plt.show()
```

```python
import matplotlib.pyplot as plt
f = open("week_1_data/dino_noisy.txt").read()
x = []
for line in f.split("\n"):
    split_line=line.split(" ")
    if len(split_line)>1:
        x.append((float(split_line[0]),float(split_line[1])))
x.append(x[0])
x = np.array(x)
saved_x = x
plt.plot(*zip(*x))
plt.show()
def smooth_curve(x, l):
    L = np.zeros((len(x),len(x)))
    for i in range(1, len(x)-1):
        L[i][i]=-2
        L[i][i-1]=1
        L[i][i+1]=1
    L[len(x)-1][0]=1
    L[len(x)-1][len(x)-1]=-2
    L[len(x)-1][len(x)-2]=1
    L[0][len(x)-1]=1
    L[0][0]=-2
    L[0][1]=1
    print(L)
    I = np.identity(len(x))
    return np.matmul((I+l*L),x)
x = saved_x
for i in range(0, 100):
    x = smooth_curve(x, 0.01)
plt.plot(*zip(*x))
plt.show()
x = saved_x
for i in range(0, 1):
    x = smooth_curve(x, 1)
plt.plot(*zip(*x))
plt.show()
x = saved_x
for i in range(0, 5):
    x = smooth_curve(x, 0.5)
plt.plot(*zip(*x))
plt.show()
```

```python
from numpy.linalg import inv
def implicit_smoothing(x, l):
    L = np.zeros((len(x),len(x)))
    for i in range(1, len(x)-1):
        L[i][i]=-2
        L[i][i-1]=1
        L[i][i+1]=1
    I = np.identity(len(x))
    return np.matmul(inv((I-l*L)),x)
x = saved_x
for i in range(0, 100):
    x = implicit_smoothing(x, 0.01)
plt.plot(*zip(*x))
plt.show()
x = saved_x
for i in range(0, 1):
    x = implicit_smoothing(x, 1)
plt.plot(*zip(*x))
plt.show()
x = saved_x
for i in range(0, 5):
    x = implicit_smoothing(x, 0.5)
plt.plot(*zip(*x))
plt.show()
x = saved_x
x = implicit_smoothing(x, 10)
plt.plot(*zip(*x))
plt.show()
```

```python
import math
import cv2
import math
from matplotlib import pyplot as plt
import numpy as np
from scipy import ndimage
im_clean = cv2.imread("data/noise_free_circles.png", cv2.IMREAD_GRAYSCALE)
im_clean = im_clean.astype(float)
plt.imshow(im_clean, cmap="gray")
plt.show()
im_noisy = cv2.imread("data/noisy_circles.png", cv2.IMREAD_GRAYSCALE)
im_noisy = im_noisy.astype(float)
plt.imshow(im_noisy, cmap="gray")
plt.show()
```

```python
import seaborn as sns 
def hist_seg(im_noisy, im_seg):
    total_hist = im_noisy.reshape(im_noisy.shape[0]*im_noisy.shape[1])
    flat_seg = im_seg.reshape(im_seg.shape[0]*im_seg.shape[1])
    segment_list = {70:[],130:[],190:[]}
    for element in zip(total_hist, flat_seg):
        segment_list[int(element[1])].append(int(element[0]))
    
    plt.hist(im_noisy.reshape(100*100), bins=256,stacked=True, color='black')
    plt.plot(np.histogram(segment_list[70], bins=256, range=(0,255))[0],color='green')
    plt.plot(np.histogram(segment_list[130], bins=256, range=(0,255))[0],color='blue')
    plt.plot(np.histogram(segment_list[190], bins=256, range=(0,255))[0],color='red')
    plt.show()
print(im_clean)
hist_seg(im_noisy, im_clean)
```
